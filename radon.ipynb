```python
import cupy as cp
import matplotlib.pyplot as plt
from functions import R,RT
```

# load object, the object should be complex64

```python
u = cp.load('u.npy').astype('complex64')
cp.save('u.npy'),u.real)
plt.imshow(u[128].real.get(),cmap='gray')
plt.colorbar()
plt.show()
plt.imshow(u[:,128].real.get(),cmap='gray')
plt.colorbar()
plt.show()
```

# Model data

```python

theta = cp.linspace(0,cp.pi,360).astype('float32')
rotation_axis = 128
data = R(u,theta,rotation_axis)

plt.imshow(data[:,0].real.get(),cmap='gray')
plt.title('angle 0')
plt.colorbar()
plt.show()

plt.imshow(data[:,45].real.get(),cmap='gray')
plt.title('angle 45')
plt.colorbar()
plt.show()

```

# Inversion: filter+backprojection

```python
urec = RT(data,theta,rotation_axis,'shepp') #filters: ramp,shepp,hann,parzen,none..

plt.imshow(urec[128].real.get(),cmap='gray')
plt.colorbar()
plt.show()
plt.imshow(urec[:,128].real.get(),cmap='gray')
plt.colorbar()
plt.show()
```
